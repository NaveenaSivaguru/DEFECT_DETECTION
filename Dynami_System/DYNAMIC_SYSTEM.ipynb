Calculating derivatives

```python
import numpy as np
import matplotlib.pyplot as plt


def derivatives(x, y, z, a, b, c):
    """Calculate derivatives"""
    dx = a * (y - x)
    dy = b * x - y - x * z
    dz = x * y - c * z
    return dx, dy, dz
```

Simulate path using Given equation

```python

def simulate_path(a, b, c, x0, y0, z0, dt, num_steps):
    """Simulate path using Given equation"""
    xs = np.zeros(num_steps)
    ys = np.zeros(num_steps)
    zs = np.zeros(num_steps)

    xs[0], ys[0], zs[0] = x0, y0, z0

    for i in range(1, num_steps):
        dx, dy, dz = derivatives(xs[i-1], ys[i-1], zs[i-1], a, b, c)
        xs[i] = xs[i-1] + dx * dt
        ys[i] = ys[i-1] + dy * dt
        zs[i] = zs[i-1] + dz * dt

    return xs, ys, zs
```

Plotting BEE path in 3D.

```python
def plot_bee_path(xs, ys, zs):
    """Plot Bee path in 3D."""
    fig = plt.figure()
    ax = fig.add_subplot(111, projection="3d")
    ax.plot(xs, ys, zs, lw=0.5)
    ax.set_title("Bee's Movement in 3D Space ")
    ax.set_xlabel("X Position")
    ax.set_ylabel("Y Position")
    ax.set_zlabel("Z Position")
    plt.show()
```

MAIN FUNCTION

```python
if __name__ == "__main__":
    # Parameters
    a, b, c = 10, 28, 2.667
    x0, y0, z0 = 0, 1, 1.05
    dt = 0.01
    num_steps = 5000

    # Run simulation
    xs, ys, zs = simulate_path(a, b, c, x0, y0, z0, dt, num_steps)

    # Plot result
    plot_bee_path(xs, ys, zs)
```
